```python
# ----------------------------------------
# STEP 0: Import Required Libraries
# ----------------------------------------

# pandas is used to load dataset
import pandas as pd
import numpy as np

# matplotlib is used to visualize results
import matplotlib.pyplot as plt
```

```python
# ----------------------------------------
# STEP 1: Load Dataset
# ----------------------------------------

data = pd.read_excel("Salary_Data.xlsx")

print("Dataset Preview:")
print(data.head())
```

```python

# ----------------------------------------
# STEP 2: Extract Input (X) and Output (Y)
# ----------------------------------------


X = data["Years of Exp."]
Y = data["Salary_employee"]

n = len(X)

print("Number of observations:", n)
```

```python
# ----------------------------------------
# STEP 3: Calculate Required Summations
# ----------------------------------------

# Σx
sum_x = np.sum(X)

# Σy
sum_y = np.sum(Y)

# Σx²
sum_x2 = np.sum(X * X)

# Σxy
sum_xy = np.sum(X * Y)

print("\nCalculated Summations:")
print("Sum of X:", sum_x)
print("Sum of Y:", sum_y)
print("Sum of X^2:", sum_x2)
print("Sum of XY:", sum_xy)
```

```python
# ----------------------------------------
# STEP 4: Apply Formula for b (Slope)
# ----------------------------------------

# Formula:
# b = [n(Σxy) − (Σx)(Σy)] / [n(Σx²) − (Σx)²]

numerator_b = (n * sum_xy) - (sum_x * sum_y)
denominator_b = (n * sum_x2) - (sum_x ** 2)

b = numerator_b / denominator_b
```

```python
# ----------------------------------------
# STEP 5: Apply Formula for a (Intercept)
# ----------------------------------------

# Formula:
# a = [(Σy)(Σx²) − (Σx)(Σxy)] / [n(Σx²) − (Σx)²]

numerator_a = (sum_y * sum_x2) - (sum_x * sum_xy)
denominator_a = denominator_b  # same denominator

a = numerator_a / denominator_a
```

```python
# ----------------------------------------
# STEP 6: Display Model Equation
# ----------------------------------------

print("\nLinear Regression Equation:")
print(f"Y = {a:.2f} + {b:.2f}X")
```

```python
# ----------------------------------------
# STEP 7: Make Predictions
# ----------------------------------------

# Predicted salaries using regression equation
Y_pred = a + b * X
```

```python
# ----------------------------------------
# STEP 8: Plot the Regression Line
# ----------------------------------------

plt.scatter(X, Y, color='blue', label="Actual Data")
plt.plot(X, Y_pred, color='red', label="Regression Line")

plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Linear Regression using Statistical Formula")
plt.legend()

plt.show()
```
