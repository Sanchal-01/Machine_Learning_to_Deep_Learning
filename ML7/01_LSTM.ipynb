```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Training Data

X=np.array([
[[1],[2],[3]],
[[2],[3],[4]],
[[3],[4],[5]],
[[4],[5],[6]],
[[5],[6],[7]],
[[6],[7],[8]],
[[7],[8],[9]]
],dtype=float)

y=np.array([4,5,6,7,8,9,10],dtype=float)

# Normalize

X=X/10
y=y/10

# Model

model=Sequential()

model.add(
    LSTM(
        32,
        input_shape=(3,1)
    )
)

model.add(Dense(1))

model.compile(
    optimizer='adam',
    loss='mse'
)

model.fit(
    X,
    y,
    epochs=500,
    verbose=1
)

# Test

test=np.array([[[8],[9],[10]]],dtype=float)

test=test/10

pred=model.predict(test)

# Convert back

pred=pred*10

print(pred)
```
